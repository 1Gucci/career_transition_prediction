```python
partial_R2 = {}

for predictor in train_X.columns:
    
    # Create a version of train_X without the current predictor
    X_reduced = train_X.drop(columns=[predictor])
    
    # Fit the reduced model
    lm_reduced = LinearRegression()
    lm_reduced.fit(X_reduced, train_y)
    
    # Predict using the reduced model
    y_pred_reduced = lm_reduced.predict(X_reduced)
    
    # Compute SSR and SSE for the reduced model
    SSR_reduced = np.sum((y_pred_reduced - train_y.mean()) ** 2)
    SSE_reduced = np.sum((train_y - y_pred_reduced) ** 2)
    
    # Compute partial R^2 for the predictor
    partial_R2[predictor] = (SSR_full - SSR_reduced) / SSE_reduced

# Create a dataframe to display predictors, their coefficients, and partial R^2 values
coeff_df = pd.DataFrame({
    'Predictor': train_X.columns,
    'Partial_R2': [partial_R2[p] for p in train_X.columns]
})
coeff_df
```

```python
# Calculate the total sum of partial R^2 values
total_partial_R2 = coeff_df['Partial_R2'].sum()

# Calculate the proportion of contribution for each predictor
coeff_df['Proportion of Contribution'] = (coeff_df['Partial_R2'] / total_partial_R2) * 100

#Display the model report of all variables
sorted_coeff_df = coeff_df.sort_values(by='Proportion of Contribution', ascending=False)
sorted_coeff_df[['Predictor', 'Partial_R2', 'Proportion of Contribution']]
```

```python
#table 6.6 backward elimination
def train_model(variables):
    model = LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    return AIC_score(train_y, model.predict(train_X[variables]), model)

best_model, best_variables = backward_elimination(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

```python
regressionSummary(valid_y, best_model.predict(valid_X[best_variables]))
```

"In forward selection,
we start with no predictors and then add predictors one by one. Each
predictor added is the one (among all predictors) that has the largest
contribution to R2 on top of the predictors that are already in it. The
algorithm stops when the contribution of additional predictors is not
statistically significant."

```python
# table 6.7 - forward selection 
#The initial model is the constant model - this requires special handling
# in train_model and score_model
def train_model(variables):
    if len(variables) == 0:
        return None
    model = LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    if len(variables) == 0:
        return AIC_score(train_y, [train_y.mean()] * len(train_y), model, df=1)
    return AIC_score(train_y, model.predict(train_X[variables]), model)

best_model, best_variables = forward_selection(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

```python
regressionSummary(valid_y, best_model.predict(valid_X[best_variables]))
```

"Stepwise regression is like forward selection except that
at each step, we consider dropping predictors that are not statistically
significant, as in backward elimination."

```python
best_model, best_variables = stepwise_selection(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

```python
regressionSummary(valid_y, best_model.predict(valid_X[best_variables]))
```

```python
# Print p-values of all predictors

dfP = pd.concat([train_X, train_y], axis=1)

# Define the formula for linear regression
predictors = train_X.columns
formula = 'TransitionYrs ~ ' + ' + '.join(predictors)

# Perform linear regression
lmB = sm.ols(formula=formula, data=dfP).fit()

# Get p-values from the summary and create a DataFrame
p_values = pd.DataFrame({'Predictor': lmB.params.index, 'P-Value': lmB.pvalues})

# Filter out the intercept
p_values = p_values[p_values['Predictor'] != 'Intercept']

# Sort the DataFrame by p-values in ascending order
sorted_p_values = p_values.sort_values(by='P-Value')

# Format p-values as decimals
sorted_p_values['P-Value'] = sorted_p_values['P-Value'].apply(lambda x: '{:.6f}'.format(x))

# Print the sorted p-values
print(sorted_p_values)
```
